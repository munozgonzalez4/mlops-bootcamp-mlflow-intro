# MLflow Tracking Server

```python
# MLflow server is running locally due to having put in the terminal: mlflow ui
import mlflow
```

```python
# Set the tracking URI to the local MLflow server
mlflow.set_tracking_uri("http://127.0.0.1:5000")
```

```python
# Creating an experiment and a run (automatic name will be generated for the run) with some metrics

mlflow.set_experiment("Check localhost connection")

with mlflow.start_run():
    mlflow.log_metric("test_1", 5)
    mlflow.log_metric("test_2", 4)

# Every time we run this cell, a new run will be created under the same experiment, and we will be able to see the metrics in the MLflow UI
```
